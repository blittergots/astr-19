```python
import numpy as np
```

*this function is to calculate the sin of x*

```python
x = (1)
print(f"np.sin({x}) = {np.sin(x)}")
```

*this function is to calculate the cosine of x*

```python
x = 1
print(f"np.cos({x}) = {np.cos(x)}")
```

*this creates a table with the sin and cos of x between 0 and 2pi.*

```python
#defining 2 pi into something readable for me
TWO_PI = 2*np.pi

x = 0 
table_text = "| sin(x) 	| cos(x) 	|"
i = 0 #this sets the increments by which we're going up
while (i<1000):
	table_text = table_text + "\n|" + str(np.sin(x)) + " 	| " + str(np.cos(x)) + " 	|"
	x = x + TWO_PI/999
	i = i + 1

print(table_text)
```
